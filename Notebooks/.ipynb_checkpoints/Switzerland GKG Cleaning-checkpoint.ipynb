```python
import pandas as pd
import numpy as np
import os

# Importing the script with all the methods definition
import GKG
import importlib

importlib.reload(GKG)
```

```python
filepath = '../GKG Notebooks Tunisia/'

column_names = ["GKGRECORDID", "V2.1DATE", "V2SOURCECOLLECTIONIDENTIFIER", "V2SOURCECOMMONNAME", "V2DOCUMENTIDENTIFIER",
                       "V1COUNTS", "V2.1COUNTS", "V1THEMES", "V2ENHANCEDTHEMES", "V1LOCATIONS", "V2ENHANCEDLOCATIONS", "V1PERSONS",
                       "V2ENHANCEDPERSONS", "V1ORGANIZATIONS", "V2ENHANCEDORGANIZATIONS", "V1.5TONE", "V2.1ENHANCEDDATES", "V2GCAM",
                       "V2.1SHARINGIMAGE", "V2.1RELATEDIMAGES", "V2.1SOCIALIMAGEEMBEDS", "V2.1SOCIALVIDEOEMBEDS", "V2.1QUOTATIONS",
                       "V2.1ALLNAMES", "V2.1AMOUNTS", "V2.1TRANSLATIONINFO", "V2EXTRASXML"]

switzerland_actors = pd.DataFrame(columns=column_names)

# Definition of the actors involved in the Swiss vote

referendum_commitee = ['Amaudruz', 'Baer Hill', 'De Weck', 'Fluckiger', 'Addor', 'Buchler', 'Filippini', 'Salzmann']
federal_council = ['Maurer', 'Cassis', 'Berset', 'Keller-Sutter', 'Amherd', 'Parmelin', 'Sommaruga', 'Thurnherr']
gun_lobby_groups = ['PROTELL']

themes_switzerland = ['ARMEDCONFLICT', 'BORDER', 'CORRUPTION', 'DEMOCRACY', 'ELECTION_FRAUD', 'EXTREMISM',
                      'FIREARM_OWNERSHIP', 'HATE_SPEECH', 'IMMIGRATION', 'INFO_HOAX', 'INFO_RUMOR',
                      'MEDIA_CENSORSHIP', 'MEDIA_SOCIAL', 'MIL_SELF_IDENTIFIED_ARMS_DEAL', 'MIL_WEAPONS_PROLIFERATION',
                      'MILITARY', 'MILITARY_COOPERATION', 'PEACEKEEPING', 'PROPAGANDA', 'SANCTIONS', 'SECURITY SERVICES',
                      'SELF_IDENTIFIED_HUMAN_RIGHTS', 'SLFID_MILITARY_SPENDING', 'SOC_INTELSHARING', 'SOVEREIGNTY',
                      'TAX_MILITARY_TITLE', 'TAX_POLITICAL_PARTY', 'TAX_WEAPONS', 'UNREST_CLOSINGBORDER']

file_score = 0
notebook_score = 0
for filename in os.listdir(filepath):
    
    if filename.endswith('.csv'):
        #print("filename: ", filename, file_score)
        
        if np.mod(file_score, 100) == 0 or file_score == 0:
            print("Score: ", file_score, notebook_score)
            
        file_score += 1
    
        path = os.path.join(filepath, filename)

        switzerland_gkg = pd.read_csv(path, sep='\t', delimiter='\t', decimal = '.', error_bad_lines=False, 
                                  header=None, engine='c', encoding='latin-1')
        switzerland_gkg.columns = column_names

        for index, row in switzerland_gkg.iterrows():
            actor_list = row['V2ENHANCEDPERSONS']
            location_list = row['V2ENHANCEDLOCATIONS']
            theme_list = row['V2ENHANCEDTHEMES']
            article_url = row['V2DOCUMENTIDENTIFIER']
            allnames_list = row['V2.1ALLNAMES']
            
            article_website = row['V2SOURCECOMMONNAME']   
                    
            if (not isinstance(article_website, float) and article_website is not None):
                if article_website[-2:] == 'ch':
                    switzerland_actors = switzerland_actors.append(row)
                    notebook_score += 1
                    #print("ch found: ", article_url)
                    
```

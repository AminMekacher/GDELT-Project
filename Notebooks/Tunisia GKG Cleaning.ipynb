```python
import pandas as pd
import numpy as np
import os

# Importing the notebook with all the methods definition
import GKG
```

```python
'''
Goal of this notebook: process the GKG Database from 09/24/2018 12pm to 09/25/2018 12pm to save in CSV files the news related to
- Principal actors in Tunisia: Essebsi, Ben Ali
- Main organizations: TBD
'''

filepath = '../GKG Notebooks Tunisia/'
    
column_names = ["GKGRECORDID", "V2.1DATE", "V2SOURCECOLLECTIONIDENTIFIER", "V2SOURCECOMMONNAME", "V2DOCUMENTIDENTIFIER",
                       "V1COUNTS", "V2.1COUNTS", "V1THEMES", "V2ENHANCEDTHEMES", "V1LOCATIONS", "V2ENHANCEDLOCATIONS", "V1PERSONS",
                       "V2ENHANCEDPERSONS", "V1ORGANIZATIONS", "V2ENHANCEDORGANIZATIONS", "V1.5TONE", "V2.1ENHANCEDDATES", "V2GCAM",
                       "V2.1SHARINGIMAGE", "V2.1RELATEDIMAGES", "V2.1SOCIALIMAGEEMBEDS", "V2.1SOCIALVIDEOEMBEDS", "V2.1QUOTATIONS",
                       "V2.1ALLNAMES", "V2.1AMOUNTS", "V2.1TRANSLATIONINFO", "V2EXTRASXML"]

tunisia_actors = pd.DataFrame(columns=column_names)
tunisia_organizations = pd.DataFrame(columns=column_names)
tunisia_country = pd.DataFrame(columns=column_names)

essebsi_score = 0
tunisia_score = 0

file_score = 0
for filename in os.listdir(filepath):
    
    if filename.endswith('.csv'):
        #print("filename: ", filename, file_score)
        
        if np.mod(file_score, 100) == 0 or file_score == 0:
            print("Score: ", file_score)
            
        file_score += 1
    
        path = os.path.join(filepath, filename)

        tunisia_gkg = pd.read_csv(path, sep='\t', delimiter='\t', decimal = '.', error_bad_lines=False, 
                                  header=None, engine='c', encoding='latin-1')
        tunisia_gkg.columns = column_names


        for index, row in tunisia_gkg.iterrows():
            if (not isinstance(row['V2ENHANCEDPERSONS'], float) and row['V2ENHANCEDPERSONS'] is not None):
                if 'Essebsi' in row['V2ENHANCEDPERSONS']:
                    tunisia_actors = tunisia_actors.append(row)
                    #print("New Ess: ", index, row['V2ENHANCEDPERSONS'])
                elif 'Ben Ali' in row['V2ENHANCEDPERSONS']:
                    tunisia_actors = tunisia_actors.append(row)                    
                elif 'Trabelsi' in row['V2ENHANCEDPERSONS']:
                    tunisia_actors = tunisia_actors.append(row)
                elif 'Chahed' in row['V2ENHANCEDPERSONS']:
                    tunisia_actors = tunisia_actors.append(row)
                    #print("New Cha: ", index, row['V2ENHANCEDPERSONS'])
                elif 'Ghannouchi' in row['V2ENHANCEDPERSONS']:
                    tunisia_actors = tunisia_actors.append(row)
                    #print("New Gha: ", index, row['V2ENHANCEDPERSONS'])
                elif 'Kanzler' in row['V2ENHANCEDPERSONS']:
                    tunisia_actors = tunisia_actors.append(row)
                elif 'Belkadhi' in row['V2ENHANCEDPERSONS']:
                    tunisia_actors = tunisia_actors.append(row)
                    
            if (not isinstance(row['V2ENHANCEDLOCATIONS'], float) and row['V2ENHANCEDLOCATIONS'] is not None):
                if 'Tunisia' in row['V2ENHANCEDLOCATIONS']:
                    tunisia_country = tunisia_country.append(row)
                    
            if (not isinstance(row['V2ENHANCEDORGANIZATIONS'], float) and row['V2ENHANCEDORGANIZATIONS'] is not None):
                if 'Ennahda' in row['V2ENHANCEDORGANIZATIONS']: 
                    tunisia_organizations = tunisia_organizations.append(row)
                elif 'Party Call' in row['V2ENHANCEDORGANIZATIONS'] or 'Party Call For Tunisia ' in row['V2ENHANCEDORGANIZATIONS']:
                    tunisia_organizations = tunisia_organizations.append(row)
```

```python
def appendDFToCSV(df, csvFilePath, sep="\t"):
    import os
    if not os.path.isfile(csvFilePath):
        df.to_csv(csvFilePath, mode='a', index=False, sep=sep)
    elif len(df.columns) != len(pd.read_csv(csvFilePath, nrows=1, sep=sep).columns):
        raise Exception("Columns do not match!! Dataframe has " + str(len(df.columns)) + " columns. CSV file has " + str(len(pd.read_csv(csvFilePath, nrows=1, sep=sep).columns)) + " columns.")
    elif not (df.columns == pd.read_csv(csvFilePath, nrows=1, sep=sep).columns).all():
        raise Exception("Columns and column order of dataframe and csv file do not match!!")
    else:
        df.to_csv(csvFilePath, mode='a', index=False, sep=sep, header=False)
```

```python
appendDFToCSV(tunisia_actors, 'tunisia_actors.csv')
appendDFToCSV(tunisia_organizations, 'tunisia_organizations.csv')
appendDFToCSV(tunisia_country, 'tunisia_country.csv')
```
